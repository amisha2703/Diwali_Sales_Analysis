# Exploratory Data Analysis

```python
# plotting a bar chart for gender vs total amount

plt.figure(figsize=(7,5))
sns.countplot(x='Gender', data=df)
plt.title('Diwali Sales',fontsize=10)
plt.xlabel('Gender',fontsize=10)
plt.ylabel('Amount',fontsize=10)
plt.show() 

```

```python
# Total Amount vs Age Group
plt.figure(figsize=(7,5))
sns.countplot(x = 'Age Group', hue= 'Gender',data = df)
plt.title('Diwali Sales',fontsize=10)
plt.xlabel('Age Group',fontsize=10)
plt.ylabel('Amount',fontsize=10)
plt.show()
```

Conclusion:

        Most of the busyers are females and even purchasing power of females are greater than men

```python
sales_state = df.groupby(['State'], as_index=False)['Orders'].sum().sort_values(by='Orders', ascending=False).head(10)

plt.figure(figsize=(15, 5))
sns.barplot(x='State', y='Orders', data=sales_state, hue='State', palette='viridis')  # Use a palette for colors
plt.title('Diwali Sales', fontsize=10)
plt.xlabel('State', fontsize=10)
plt.ylabel('Orders', fontsize=10)
plt.xticks(rotation=45)
plt.show()
```

```python
sales_state = df.groupby(['State'], as_index=False)['Amount'].sum().sort_values(by='Amount', ascending=False).head(10)

plt.figure(figsize=(15, 5))
sns.barplot(x='State', y='Amount', data=sales_state, hue='State',palette='coolwarm')
plt.title('Diwali Sales', fontsize=10)
plt.xlabel('State', fontsize=10)
plt.ylabel('Amount', fontsize=10)
plt.xticks(rotation=45)
plt.show()
```

Conclusion:

  most of the orders & total sales/amount are from Uttar Pradesh, Maharashtra and Karnataka respectively

```python

plt.figure(figsize=(7, 5))
ax=sns.countplot(x='Marital_Status', data=df, palette='pastel')

plt.title('Diwali Sales', fontsize=10)
plt.xlabel('Marital Status', fontsize=10)
plt.ylabel('count', fontsize=10)

for container in ax.containers:
    ax.bar_label(container, fmt='%.0f')  # Displays the sum without decimals

plt.show()


```

```python
plt.figure(figsize=(15, 5))
ax=sns.barplot(x='Marital_Status', y='Amount', hue='Gender', data=df, estimator=sum)
plt.title('Diwali Sales', fontsize=10)
plt.xlabel('Marital Status', fontsize=10)
plt.ylabel('Total Amount', fontsize=10)

for container in ax.containers:
    ax.bar_label(container, fmt='%.0f')  # Displays the sum without decimals

plt.show() 


```

Conclusion:

*From above graphs we can see that most of the buyers are married (women) and they have high purchasing power*

```python
plt.figure(figsize=(15,5))
sns.countplot(x='Occupation', data=df, palette='tab10')
plt.title('Diwali Sales',fontsize=10)
plt.xlabel('Occupation',fontsize=10)
plt.show()
```

```python
sales_state=df.groupby(['Occupation'],as_index=False)['Amount'].sum().sort_values(by='Amount',ascending=False).head(10)

plt.figure(figsize=(15,5))
sns.barplot(x ='Occupation', y='Amount',data = sales_state, palette='viridis')
plt.title('Diwali Sales',fontsize=10)
plt.xlabel('Occupation',fontsize=10)
plt.ylabel('Amount',fontsize=10)
# plt.grid()
plt.show()
```

Conclusion:

*From above graphs we can see that most of the buyers are working in IT, Healthcare and Aviation sector*

```python
plt.figure(figsize=(15,5))
sns.countplot(x='Product_Category', data=df , color='purple')
plt.title('Diwali Sales',fontsize=10)
plt.xlabel('Product_Category',fontsize=10)
plt.xticks(rotation=45)
plt.show()
```

```python
sales_state=df.groupby(['Product_Category'],as_index=False)['Amount'].sum().sort_values(by='Amount',ascending=False).head(10)

plt.figure(figsize=(15,5))
sns.barplot(x ='Product_Category', y='Amount',data = sales_state, palette='Set2')
plt.title('Diwali Sales',fontsize=10)
plt.xlabel('Product_Category',fontsize=10)
plt.ylabel('Amount',fontsize=10)
plt.xticks(rotation=45)
plt.show()
```

Conclusion:

*From above graphs we can see that most of the sold products are from Food, Clothing and Electronics category*

```python
sales_state = df.groupby(['Product_ID'], as_index=False)['Orders'].sum().sort_values(by='Orders', ascending=False).head(10)

plt.figure(figsize=(15,5))
sns.barplot(x ='Product_ID', y='Orders',data = sales_state, palette='coolwarm')
plt.title('Diwali Sales',fontsize=10)
plt.xlabel('Product_ID',fontsize=10)
plt.ylabel('Orders',fontsize=10)
# plt.xticks(rotation=45)
plt.show()



```
